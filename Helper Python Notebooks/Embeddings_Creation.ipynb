```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
```

```python
import json

# Load JSON data
with open('C:/Users/ObelWork-1/Downloads/data_for_vectordb.json', 'r') as file:
    data = json.load(file)

# Extract descriptions and CVE IDs
descriptions = [item['description'] for item in data]
cve_ids = [item['id'] for item in data]
```

## Creation of Embeddings of Description

```python
model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
```

```python
embeddings = model.encode(descriptions, convert_to_tensor=True)
```

```python
import torch
torch.save(embeddings, 'C:/Users/ObelWork-1/Downloads/cve_embedding.pt')
```
